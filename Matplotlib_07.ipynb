```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.array([75,95,115,135,155,175,195])
c = np.array([9,19,36,46,51,55,60])
```

```python
plt.figure(figsize=(11,8))
plt.style.use('ggplot')

hfont = {'fontname':'Comic Sans MS'}

plt.plot(x,c, color='green', marker='o', label='Cumulative Frequency')
plt.fill_between(x,c, color='lightgreen', alpha=0.4)




plt.xlabel('Midpoints', fontsize=16, **hfont)
plt.ylabel('Cumulative Frequency',fontsize=16, **hfont)
plt.title('Cumulative Frequency Curve', fontsize=16, fontweight='bold',**hfont)
plt.legend()
plt.xlim(70,200)
plt.ylim(8,65)

plt.tight_layout()
plt.show()
```
