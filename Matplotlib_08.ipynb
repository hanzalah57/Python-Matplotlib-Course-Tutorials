```python
import matplotlib.pyplot as plt
```

```python
labels = ['Apple', 'Samsung', 'Xiomi', 'BBk', 'Others']
explode = (0.1,0,0,0,0)
share = [10.8,7,6,10,15]
hfont = {'fontname':'Comic Sans MS'}
```

```python
plt.figure(figsize=(8,8))
plt.style.use('ggplot')
plt.pie(x=share, labels=labels, autopct='%.2f%%', shadow=False, startangle=90)
plt.title("Share of Companies", fontsize=17, fontweight='bold', **hfont)
plt.axis('equal')
plt.legend(loc='upper right')

circle = plt.Circle(xy=(0,0), radius=0.75, facecolor='white')
plt.gca().add_artist(circle)


plt.show()
```
