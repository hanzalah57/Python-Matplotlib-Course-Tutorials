```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('gas bill.xlsx')
```

```python
df.describe()
```

```python
df.info()
```

```python
df['HM3']
```

```python
plt.style.use('ggplot')

fig, ax1 = plt.subplots(figsize=(12,6))

hfont = {'fontname':'Comic Sans MS'}

color1 = 'blue'
line1 = ax1.plot(df['Month'], df['Bill'], color=color1, label="Bill in RS")
ax1.set_ylabel('Bill in RS', color=color1, fontsize=19, fontweight='bold', **hfont)
ax1.set_xlabel('Months', color='black', fontsize=19, fontweight='bold', **hfont)
ax1.tick_params(axis='y', labelcolor=color1, labelsize=14)
ax1.tick_params(axis='x', labelrotation=35, labelcolor='black', labelsize=14)
ax1.set_title('Gas Bill Analysis', color='black', fontsize=20, fontweight='bold', **hfont)


color2 = 'crimson'
ax2 = ax1.twinx()
line2 = ax2.plot(df['Month'], df['HM3'], color=color2, label="HM3")
ax2.set_ylabel('HM3', color=color2, fontsize=19, fontweight='bold', **hfont)
ax2.tick_params(axis='y', labelcolor=color2, labelsize=14)
ax2.spines['left'].set_color(color1)
ax2.spines['right'].set_color(color2)

line = line1 + line2
labels = [l.get_label() for l in line]
ax1.legend(line,labels)


plt.tight_layout()
plt.grid(False)
plt.show()
```
