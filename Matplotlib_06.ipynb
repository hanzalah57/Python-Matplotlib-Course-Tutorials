```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
r = [106, 107, 76, 82, 109, 107, 115, 93, 187, 95, 123, 125, 111, 92, 86, 70, 126, 68, 130, 129, 139, 115, 119, 115, 128,
    100, 186, 84, 99, 113, 204, 111, 141, 136, 123, 90, 115, 98, 110, 78, 185, 162, 178, 140, 152, 173, 146, 158, 194, 148, 90,
    107, 181, 131, 75, 184, 104, 110, 80, 118, 82]
print(len(r))

bins = [65, 85, 105, 125, 145, 165, 185, 205]
```

```python
m = np.median(r)
```

```python
plt.figure(figsize=(10,7))
plt.style.use('ggplot')
plt.hist(r, bins=bins, color='#ffb3ff', edgecolor='#990099')

plt.axvline(m, color='#008ae6', label='Median Rainall')
plt.legend()

plt.title("Rainfall Distribution", fontsize=15, fontweight='bold')
plt.xlabel('Rainfall-mm', fontsize=15)
plt.ylabel('Rainfall Frequency', fontsize=15)
plt.xlim(60,210)


plt.tight_layout()
plt.show()
```
