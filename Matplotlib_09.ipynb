```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
x = np.linspace(0,100,1000)
y = np.sqrt(x)
```

```python
plt.style.use('ggplot')
fig, ax1 = plt.subplots(figsize=(7,4))

color1 = 'blue'
line1 = ax1.plot(x,x, color=color1, label="Linear")
ax1.set_ylabel('Linear', color=color1)
ax1.tick_params(axis='y', labelcolor=color1)


color2 = 'crimson'
ax2 = ax1.twinx()
line2 = ax2.plot(x,y, color=color2, label="Square Root")
ax2.set_ylabel('Square Root', color=color2)
ax2.tick_params(axis='y', labelcolor=color2)
ax2.spines['left'].set_color(color1)
ax2.spines['right'].set_color(color2)

line = line1 + line2
labels = [l.get_label() for l in line]
ax1.legend(line,labels)


plt.tight_layout()
plt.grid(False)
plt.show()
```
