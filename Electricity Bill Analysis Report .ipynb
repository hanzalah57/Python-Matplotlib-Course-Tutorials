```python
import pandas as pd 
import matplotlib.pyplot as plt  
```

```python
df = pd.read_excel('electric bill.xlsx')
data = pd.DataFrame(df)
data
```

```python
data.describe()
```

```python
data.info()
```

```python
data['Months']
```

```python
data['Units ']
```

```python
plt.figure(figsize=(13,8))
plt.style.use('bmh')

hfont = {'fontname':'Comic Sans MS'}

plt.plot(data['Months'], data['Units '], color='green', marker='o', lw=2, label='Units')

plt.fill_between(data['Months'], data['Units '], color='lightgreen', alpha=0.4)




plt.xlabel('Months', fontsize=26, **hfont)
plt.ylabel('Units Consumed',fontsize=26, **hfont)
plt.title('Electricity Bill Units Analysis', fontsize=26, fontweight='bold',**hfont)
plt.legend(loc='upper left')
plt.ylim(0,200)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13,8))
plt.style.use('bmh')

plt.plot(data['Months'], data['Bill '], color='red', marker='o', lw=2, label='Bill-RS')

plt.fill_between(data['Months'], data['Bill '], color='#ff9999', alpha=0.2)




plt.xlabel('Months', fontsize=26, **hfont)
plt.ylabel('Bill-RS',fontsize=26, **hfont)
plt.title('Electricity Bill-RS Analysis', fontsize=26, fontweight='bold',**hfont)
plt.legend()
plt.ylim(400,3000)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13,8))
plt.style.use('bmh')

hfont = {'fontname':'Comic Sans MS'}

plt.bar(data['Months'], data['Units '], width=0.9, color='#b3ffb3', edgecolor='#004d00', bottom=None, label='Units Bars')
plt.plot(data['Months'], data['Units '], color='#b3003b', marker='o', lw=3, label='Units Curve')
#plt.fill_between(data['Months'], data['Units '], color='lightgreen', alpha=0.4)




plt.xlabel('Months', fontsize=26, **hfont)
plt.ylabel('Units Consumed',fontsize=26, **hfont)
plt.title('Electricity Bill Units Analysis', fontsize=26, fontweight='bold',**hfont)
plt.legend(loc='upper left')
plt.ylim(0,200)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13,8))
plt.style.use('bmh')

plt.bar(data['Months'], data['Bill '], width=0.9, color='#ffb3cc', edgecolor='#4d0019', bottom=None, label='Bill in RS')
plt.plot(data['Months'], data['Bill '], color='#330066', marker='o', lw=3, label='Bill Curve')





plt.xlabel('Months', fontsize=26, **hfont)
plt.ylabel('Bill-RS',fontsize=26, **hfont)
plt.title('Electricity Bill-RS Analysis', fontsize=26, fontweight='bold',**hfont)
plt.legend()
plt.ylim(400,3000)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=1,nrows=2, figsize=(13,8))

ax1 = ax[0]
ax2 = ax[1]

#first plot
ax1.plot(data['Months'], data['Units '], color='green', marker='o', lw=2, label='Units')
ax1.set_xlabel('Months',fontsize=20, **hfont)
ax1.set_ylabel('Units Consumed', fontsize=20, **hfont)
ax1.set_title('Electricity Bill Units', fontsize=20, fontweight='bold', **hfont)
ax1.fill_between(data['Months'], data['Units '], color='lightgreen', alpha=0.4)
ax1.set_ylim(0,200)
ax1.legend()



#second plot
ax2.plot(data['Months'], data['Bill '], color='red', marker='o', lw=2, label='Bill-RS')
ax2.set_xlabel('Months',fontsize=20, **hfont)
ax2.set_ylabel('Electricity Bill in RS', fontsize=20, **hfont)
ax2.set_title('Electricity Bill-RS', fontsize=20, fontweight='bold', **hfont)
ax2.fill_between(data['Months'], data['Bill '], color='#ff9999', alpha=0.4)
ax2.set_ylim(400,3000)
ax2.legend()

plt.tight_layout()
plt.show()
```
