```python
import matplotlib.pyplot as plt
import matplotlib
#print(matplotlib.__version__)
import numpy as np

print(plt.style.available)

plt.style.use('fivethirtyeight')


t=np.arange(1,11)

b=np.arange(len(t))

width=0.25


v=np.array([5.1,7.8,11.2,13.9,16.7,20.1,22.9,26.7,30.3,33.1])

w=np.array([7.8,9.4,3.3,8.5,5.7,2.1,3.0,6.6,9.4,4.8])

    

plt.bar(b-width,v,color='#d09696',width=width, label='Velocity Curve')

plt.bar(b+width,w,color='#003d66',width=width, label='wrape')

#plt.plot(t,v,color='#003d66',linewidth='3', label='Velocity Curve')


plt.xticks(ticks=b,labels=t)

plt.title("Velocity Plot")
plt.xlabel("Time(sec)")
plt.ylabel("F(t)")
plt.legend(loc='upper left')
#plt.grid(True)



#comic style
#plt.xkcd()


#to reverse comic effect
#plt.rcdefaults()


plt.tight_layout()


#plt.savefig('plot.png')


plt.show()

```
