```python
import matplotlib.pyplot as plt
import numpy as np

print(plt.style.available)

plt.style.use("fivethirtyeight")

t1=np.arange(1,10,0.5)
t2=np.arange(1,10,0.5)

e1=np.exp(-t1)
e2=np.exp(t2)

plt.plot(t1,e1,color='b',label='Decay Curve')

#plt.subplot(t2,e2,color='g',marker='o',label='Growth Curve')

plt.tight_layout()
plt.legend()
plt.title('Exponential Function')
plt.xlabel('t')
plt.ylabel('f(t)')

plt.show()
```
