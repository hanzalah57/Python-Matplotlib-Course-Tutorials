```python
import matplotlib.pyplot as plt

print(plt.style.available)
plt.style.use('_classic_test_patch')


u=[40,52,85,38]
labels=['40mm',"52mm","85mm","38mm"]
colors=['#c6743d', '#523d50','#a1b3a8','#8ab3cd']

#exploding slice
explode=[0,0,0.1,0]




plt.pie(u,labels=labels,colors=colors,explode=explode,
        shadow=True,startangle=47,autopct='%1.1f%%',
        wedgeprops={'edgecolor':'white'})




#plt.legend()
plt.title('Last Four Months Islamabad Rainfall')
plt.tight_layout()

plt.show()
```
